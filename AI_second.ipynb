```python
def Findmax(list1):
    max = 0
    for i in range(0, len(list1)):
        if (list1[i] > max):
            max = list1[i]
    return max

list1 = [1,2,3,4,5,6,8,23,45]
print(Findmax(list1))
list2 = [45,22,66,77,88]
print(Findmax(list2))
```

```python
from math import sqrt

def Pythagoras(base, height):
    ###
    # 이곳에 빗변의 길이를 리턴하는 함수를 구현하시오.
    hypotenuse = sqrt(base ** 2 + height ** 2)
    
    # 빗변은 영어로 hypotenuse이다.
    ###
    return hypotenuse

print(Pythagoras(5, 2))
```

연습 문제3

정수를 1개 입력 받아 1부터 입력 받은 수까지 각각에 대해 제곱을 구한 후 출력하는 프로그램을 작성하시오.

음수인 정수, 0이 입력으로 들어오면

 "음수는 자연수가 아닙니다."
 
 "0은 1보다 작습니다."

 를 출력한 후 프로그램을 종료하세요.

(종료 방법은 break를 사용하기)

(while문을 사용하세요)

```python
# 연습문제 3 답안
a = int(input("정수를 입력해주세요.: "))
now = 1
while (now !=  a + 1):
    if (a <= 0 ):
        print("음수는 자연수가 아닙니다.")
        print("0은 1보다 작습니다.")
        break
    print(now * now)
    now = now + 1
```
