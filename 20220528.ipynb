**## 조건문**

if, elif, else 등이 있다.
만약에라는 뜻을 가진 영어와 비슷하게 If 이후에 오는 조건이 참이면 아래 구문을 실행하는 성격을 띈다.

만약 달면:
    삼킨다.
그렇지 않으면:

한글로 구현한 if 문이다. 달면 삼키고 달지 않다(쓰다, 맵다, 짜다. 등)면 뱉는다. 가 실행된다.

```python
a = 1
if a == 1:
    print('a는 1이다.')
else:
    print('a는 1이 아니다.')
```

:을 기준으로 구문들을 구분한다. 위 예제의 경우
if a == 1: 다음에 오는 내용들이 조건문의 구문이 된다.

단, 조건문의 구문으로 넣어줄 경우 if 보다 바깥쪽에 있어야 한다. 들여쓰기라고 부른다. 들여쓰기가 되어 있는 line만 조건문의 구문이라고 볼 수 있다.
주로 tab을 한 번 쳐서 들여쓰기를 해준다.

조건문 뿐만이 아닌 이후에 나올 반복문, 함수 등에서도 :을 사용하고 들여쓰기를 해줘야 한다.

```python
a = 2
if a == 1:                      # a = 1과 a == 1은 다른 표현이다. a = 1이라는 표현은 a에 1을 저장한다라는 뜻
                                # a == 1은 a와 1을 비교해서 같으면 True 틀리면 False를 반환한다.
    print('a는 1이다.')
else:
    print('a는 1이 아니다.')
```

if (조건) 조건에서 사용되는 연산자들은 다양하다.

우리는 조건문과 반복문을 학습했다. 조건문과 반복문은 함께 사용할 수 있다. 아래 나오는 예시들을 살펴보자


```python
for i in range(0, len(department)):
    if department[i] == '컴퓨터공학과 좋아!':
        print("정말?")
    elif department[i] == '산업공학과 좋아!':
        print("그렇구나!")
    else:
        print("멘토는 컴퓨터공학이 더 좋아!")
```

반복문은 반복을 제어할 수 있습니다.
break를 걸면 반복문이 아예 멈추고 반복문 다음에 있는 구문이 실행됩니다.

continue를 넣을 경우 반복문 안에서 해당 라인에서 멈추고 다음 반복을 시작합니다.

```python
a = 500
while a > 0:
    print(a)
    a -= 1
    continue                            # continue 아래에 있는 내용들은 무시된다!
    if a == 450:
        print("450에 도달 했습니다!")
    if a == 400:
        print("400에 도달 했습니다!")
    if a == 380:
        print("380에 도달 했는데 더 이상 숫자를 보기 싫습니다.")
        print("작동을 멈춥니다!")
        break
print('while밖은 정상적으로 출력이 된다!')
```
