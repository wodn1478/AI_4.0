리스트라고 부르는 자료형

```python
cafe = ['Twosome', 'starbucks', 'ediya', '빽다방']
print(cafe)
```

cafe라고 하는 변수에 카페 리스트가 담긴 모습을 볼 수 있다.
왜 리스트라는 자료형이 필요할까?
데이터들을 쉽게 관리하고 불러올 수 있기 때문이다. 또, 비슷한 특성(카페라는 공통점)을 가진 것들을 함께 관리할 수 있기 때문이다.

```python
len_cafe = len(cafe)
print(len_cafe)
print(len(cafe))
```

```python
print(cafe[0], cafe[1], cafe[2], cafe[3])
```

리스트는 0번부터 시작한다. 1번부터 시작한다고 생각하는 경우가 종종 있다. 코딩을 자주하다 보면 익숙해지지만 자주 실수하는 부분이다.

remove(a) a를 리스트에서 제거한다.
외에도 insert, reverse, clear 등을 사용할 수 있다.
아래는 그 예이다. 


```python
cafe.remove('빽다방')
print(cafe)
```

```python
cafe.append('MEGA')
print(cafe)
```

```python
cafe.append('MEGA')
print(cafe)
```

MEGA가 중복 되었다. 우리는 맨 끝의 MEGA를 지우지 않고 다른 카페 이름으로 변경하려고 한다. 어떻게 해야할까?


```python
cafe[4] = 'venti'
print(cafe)
```

만약에 중복된 것을 알고 있는데 리스트가 너무 길어서 몇 번째인지 모르는 상황이면 가장 간단하게 첫 번째를 변경할 수 있다.


```python
cafe[4] = 'MEGA'
print(cafe)
cafe[cafe.index('MEGA', 0, len(cafe))] = 'venti'
print(cafe)
```

**## 조건문**

if, elif, else 등이 있다.
만약에라는 뜻을 가진 영어와 비슷하게 If 이후에 오는 조건이 참이면 아래 구문을 실행하는 성격을 띈다.

만약 달면:
    삼킨다.
그렇지 않으면:
    뱉는다.

한글로 구현한 if 문이다. 달면 삼키고 달지 않다(쓰다, 맵다, 짜다. 등)면 뱉는다. 가 실행된다.

```python
a = 1
if a == 1:
    print('a는 1이다.')
else:
    print('a는 1이 아니다.')
```

```python
a = 2
if a == 1:                      # a = 1과 a == 1은 다른 표현이다. a = 1이라는 표현은 a에 1을 저장한다라는 뜻
                                # a == 1은 a와 1을 비교해서 같으면 True 틀리면 False를 반환한다.
    print('a는 1이다.')
else:
    print('a는 1이 아니다.')
```

if (조건) 조건에서 사용되는 연산자들은 다양하다.
파이썬의 경우
==, !=, >, <, >=, <=, and, or, () 등이 있다.

아래는 조건을 다양하게 해서 표현한 if문이다.

```python
if (True):
    print("여기는 A")
else:
    print("여기는 B")

if (True and False):
    print("여기는 C")
else:
    print("여기는 D")

if (True or False):
    print("여기는 E")
else:
    print("여기는 F")

if ( 4 > 3):
    print("여기는 G")
if ( 4 >= 4):
    print("여기는 H")
if ( 4 != 3):
    print("여기는 I")
if (4 != 3 and 4 > 3 and 5 > 3 and 5 == 5):
    print("여기는 J")

if (0):
    print("0은 조건에서 False와 같다 이 내용은 출력되지 않는다.")
elif (3):
    print("3은 조건에서 True와 같다. 0을 제외한 모든 정수는 True와 같다.")
else:
    print("위에서 통과되는 조건이 없는 경우 모두 이 내용이 출력된다.")


```

유니스트 편의점에서는 사과를 1개에 500원, 바나나를 1개에 700원에 팔고 있다. 사과 5개, 바나나 3개를 사고 5000원 지폐로 결제할 경우 거스름 돈이 남을 경우 거스름 돈을 출력하고, 남지 않을 경우 "거스름 돈이 남지 않음."을 출력한다.

if 문을 이용해서 구현할 것, 사과와 바나나의 가격을 변수에 담고 거스름 돈 역시 변수에 담을 것.

```python
apple = 500
banana = 700
price = apple * 5 + banana * 3
change = 5000 - price
if (change):
    print(change)
else:
    print("거스름 돈이 남지 않음.")
```

반복문
while과 for 문이 있다.

반복문은 조건에 통과된다면 해당 반복문에 들어있는 구문들을 반복하여 실행한다.

while은
while(조건):
    구문1
    구문2
    구문3...
처럼 표현된다.


for은
for i in (순서):
    구문1
    구문2
    구문3...
처럼 표현된다.

while은 조건에 통과만 하면 반복할 수 있고, for은 순서가 있는 리스트나 튜플을 활용할 때 자주 사용된다. 또한 i를 변수처럼 사용할 수 있기 때문에 indexing(순서 매기기)를 할 때도 자주 활용된다.
